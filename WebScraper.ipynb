```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'

# header due to Wikipedia requiring a User-Agent to be stated per their robot policy

header = {
    'User-Agent': 'WebScraperProject/1.0 (Texas State Univeristy Personal Project; pow10@txstate.edu) Python/3.12.10 requests/2.32.5'
}

page = requests.get(url, headers=header)

soup = BeautifulSoup(page.text, 'html.parser')

# print(soup)
```

```python

# wikitable sortable sticky-header-multi sort-under

table = soup.find('table',class_='wikitable sortable sticky-header-multi sort-under')
```

```python
world_titles = table.find_all('th')
# first 9 elements are the column titles of the table
# row titles also use the th element. 
world_table_titles = [title.text.strip() for title in world_titles[:7]] 
table_col_len = len(world_table_titles)
# overriding the last column title to remove unecessary text
world_table_titles[table_col_len - 1] = world_table_titles[table_col_len - 1][:12] 
print(world_table_titles)
```

```python
import pandas as pd

df = pd.DataFrame(columns = world_table_titles)
df
```

```python
table_column = table.find_all('tr')
```

```python
# Data processing 
rank = 1
for row in table_column[2:]:
    row_data = row.find_all('td')
    # print(row_data)
    individual_row_data = [data.text.strip() for data in row_data]
    individual_row_data.insert(0, str(rank))
    # print(individual_row_data[:7]) #for testing
    rank += 1
    length = len(df)
    df.loc[length] = individual_row_data[:7]
```
