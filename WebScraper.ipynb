```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'

# header due to Wikipedia requiring a User-Agent to be stated per their robot policy

header = {
    'User-Agent': 'WebScraperProject/1.0 (Texas State Univeristy Personal Project; pow10@txstate.edu) Python/3.12.10 requests/2.32.5'
}

page = requests.get(url, headers=header)

soup = BeautifulSoup(page.text, 'html.parser')

# print(soup)
```

```python

# wikitable sortable sticky-header-multi sort-under

table = soup.find('table',class_='wikitable sortable sticky-header-multi sort-under')
```

```python
world_titles = soup.find_all('th')

world_titles
```
